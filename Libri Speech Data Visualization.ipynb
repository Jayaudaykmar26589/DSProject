```python
from audio_dataset import *
sr = 8000
duration = 5
voice_ds = AudioDataSet('./../LibriSpeech/dev-clean', sr=sr, duration=duration)
```

```python
from common_audio import *
male = voice_ds.load([0])
male_spectrogram = audio_to_spectrogram(male, sr)
male_mel_spectrogram = audio_to_spectrogram(male, sr, mel=True)
female = voice_ds.load([41])
female_spectrogram = audio_to_spectrogram(female, sr)
female_mel_spectrogram = audio_to_spectrogram(female, sr, mel=True)
```

```python
fig = plt.figure(figsize=(15,10))
fig.subplots_adjust(hspace=0.2, wspace=0.4)

fig.add_subplot(2, 1, 1)
show_audio(male[0],sr)
plt.title('Male Voice')

fig.add_subplot(2, 1, 2)
show_audio(female[0],sr)
plt.title('Female Voice')
```

**These are what the mel spectrograms look like**

```python
fig = plt.figure(figsize=(20,15))
fig.subplots_adjust(hspace=0.2, wspace=0.4)

fig.add_subplot(2, 2, 1)
librosa.display.specshow(male_mel_spectrogram[0], sr=sr, hop_length=64, x_axis='time', y_axis='hz')
plt.title("Male Voice in linear scale")
plt.tight_layout() 


fig.add_subplot(2, 2, 2)
librosa.display.specshow(male_mel_spectrogram[0], sr=sr, hop_length=64, x_axis='time', y_axis='log')
plt.title("Male Voice in log scale")
plt.tight_layout() 

fig.add_subplot(2, 2, 3)
librosa.display.specshow(female_mel_spectrogram[0], sr=sr, hop_length=64, x_axis='time', y_axis='hz')
plt.title("Female Voice in linear scale")
plt.tight_layout() 

fig.add_subplot(2, 2, 4)
librosa.display.specshow(female_mel_spectrogram[0], sr=sr, hop_length=64, x_axis='time', y_axis='log')
plt.title("Female Voice in log scale")
plt.tight_layout() 
```

```python
fig = plt.figure(figsize=(20,15))
fig.subplots_adjust(hspace=0.2, wspace=0.4)

fig.add_subplot(2, 2, 1)
librosa.display.specshow(male_spectrogram[0], sr=sr, hop_length=64, x_axis='time', y_axis='hz')
plt.title("Male Voice in linear scale")
plt.tight_layout() 


fig.add_subplot(2, 2, 2)
librosa.display.specshow(male_spectrogram[0], sr=sr, hop_length=64, x_axis='time', y_axis='log')
plt.title("Male Voice in log scale")
plt.tight_layout() 

fig.add_subplot(2, 2, 3)
librosa.display.specshow(female_spectrogram[0], sr=sr, hop_length=64, x_axis='time', y_axis='hz')
plt.title("Female Voice in linear scale")
plt.tight_layout() 

fig.add_subplot(2, 2, 4)
librosa.display.specshow(female_spectrogram[0], sr=sr, hop_length=64, x_axis='time', y_axis='log')
plt.title("Female Voice in log scale")
plt.tight_layout() 
```

**The above log scale graphs basically stretched the spectrograms: the bottom graphs will plot the log transformed spectrogram in the linear scale**

```python
male_log_channels = complex_to_log_magnitude_phase_channels(male_spectrogram)
male_real, male_imag = male_log_channels[0,0], male_log_channels[0,1]
female_log_channels = complex_to_log_magnitude_phase_channels(female_spectrogram)
female_real, female_imag = female_log_channels[0,0], female_log_channels[0,1]




fig = plt.figure(figsize=(20,15))
fig.subplots_adjust(hspace=0.2, wspace=0.4)

fig.add_subplot(2, 2, 1)
librosa.display.specshow(male_real, sr=sr, hop_length=64, x_axis='time', y_axis='hz')
plt.title("Male Voice in linear scale")
plt.tight_layout() 


fig.add_subplot(2, 2, 2)
librosa.display.specshow(male_imag, sr=sr, hop_length=64, x_axis='time', y_axis='hz')
plt.title("Male Voice in log scale")
plt.tight_layout() 

fig.add_subplot(2, 2, 3)
librosa.display.specshow(female_real, sr=sr, hop_length=64, x_axis='time', y_axis='hz')
plt.title("Female Voice in linear scale")
plt.tight_layout() 

fig.add_subplot(2, 2, 4)
librosa.display.specshow(female_imag, sr=sr, hop_length=64, x_axis='time', y_axis='hz')
plt.title("Female Voice in log scale")
plt.tight_layout() 




```
